## Clasificación supervisada

Para un modelo de clasificación supervisado con el dataset que nos han dado el objetivo sera clacificar el hongo con las clases que nos han dado para cumplir con ello se usarán las características que nos dan (la forma de la capa, la superficie de la capa, el color de la branquia, etc.) usando las clases la red neuronál se corregirá cuando compare con la respuesta correcta.

Debido a que los datos estan como letras, para que la red pueda ser entrenada se hará un preprocesamiento de los datos y convertir las letras a numeros que es con lo que la red se puede entrenar, despues se procederá con el entrenamiento, por ultimo se hace un pequeño test para comprobar su efectividad, en un tamaño de prueba de 10 datos se obtuvo 100% de efectividad.


```python
#Importar librerías
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
datos
```

```python
dataset = datos
for columna in dataset.columns:
    le = LabelEncoder()
    le.fit(dataset[columna])
    le.classes_
    dataset[columna] = le.transform(dataset[columna])
```

```python
dataset.shape
```

```python
#Importar librerías
import numpy
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = numpy.array(datos[1:])
```

```python
X = dataset[:,1:]
y = dataset[:,0]
#print(y)
```

```python
#definir el modelo, será secuencial dado que se pondrá un layer tras otro totalmente conectados
model = Sequential()
#Capa de entrada y primer capa oculta(ambos se ejecutan con esta instrucción)
model.add(Dense(12, input_dim=22, activation='relu'))
#Segunda capa oculta
model.add(Dense(22, activation='relu'))
#Capa de salida
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=150, batch_size=10)
```

```python
predictions = model.predict_classes(X)
for i in range(10):
    print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```
